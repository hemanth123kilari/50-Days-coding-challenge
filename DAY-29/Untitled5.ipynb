```python
# Input Data
students = {
    "Asha": {"Math": 78, "Science": 85, "English": 67},
    "Ravi": {"Math": 35, "Science": 40, "English": 50},
    "John": {"Math": 90, "Science": 92, "English": 88}
}

results = {}

for name, marks in students.items():
    total = sum(marks.values())
    average = total / len(marks)

    results[name] = {
        "Total": total,
        "Average": average
    }

print("Student Results:")
for name, data in results.items():
    print(name, "-> Total:", data["Total"], ", Average:", round(data["Average"], 2))

```

```python
# TASK2
top_scorer = max(results, key=lambda x: results[x]["Total"])
print("\nTop Scorer:", top_scorer)
```

```python
#TASK-3
failed_students = []

for name, marks in students.items():
    for subject, score in marks.items():
        if score < 40:
            failed_students.append(name)
            break   # stop checking further subjects

print("\nFailed Students:", failed_students)
```
