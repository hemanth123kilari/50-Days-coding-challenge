```python
nested = [[1, 2], [3, 4], [5]]

flat = [num for sublist in nested for num in sublist]

print(flat)



```

```python
s1 = "listen"
s2 = "silent"

result = sorted(s1) == sorted(s2)

print(result)
```

```python
nums = [1, 2, 4, 5]

full_set = set(range(1, 6))   # 1 to 5
missing = list(full_set - set(nums))[0]

print(missing)
```

```python
words = ["apple", "ant", "banana", "bat"]

grouped = {}

for word in words:
    key = word[0]
    grouped.setdefault(key, []).append(word)

print(grouped)
```

```python
data = {"a": 1, "b": None, "c": 3, "d": None}

cleaned = {k: v for k, v in data.items() if v is not None}

print(cleaned)
```
