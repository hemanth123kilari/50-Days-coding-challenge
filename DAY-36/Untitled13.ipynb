```python
def calculate_average(values):
    if len(values) == 0:
        return 0
    return sum(values) / len(values)
```

```python
def find_top_performers(data, column, threshold):
    result = []
    for student in data:
        if student[column] >= threshold:
            result.append(student["name"])
    return result
```

```python
def find_at_risk_students(data, column, threshold):
    result = []
    for student in data:
        if student[column] < threshold:
            result.append(student["name"])
    return result
```

```python
def batch_summary(data):
    batches = {}

    for student in data:
        batch = student["batch"]

        if batch not in batches:
            batches[batch] = {
                "consumption": [],
                "attendance": [],
                "assignment": []
            }

        batches[batch]["consumption"].append(student["consumption"])
        batches[batch]["attendance"].append(student["live"])
        batches[batch]["assignment"].append(student["assignment_sub"])

    summary = {}

    for batch in batches:
        summary[batch] = {
            "avg_consumption": calculate_average(batches[batch]["consumption"]),
            "avg_attendance": calculate_average(batches[batch]["attendance"]),
            "avg_assignment": calculate_average(batches[batch]["assignment"])
        }

    return summary
```

```python
def rank_batches(batch_summary_data):
    best_batch = None
    best_score = 0

    for batch in batch_summary_data:
        data = batch_summary_data[batch]

        # overall score = average of all metrics
        score = (data["avg_consumption"] + data["avg_attendance"] + data["avg_assignment"]) / 3

        if score > best_score:
            best_score = score
            best_batch = batch

    return best_batch
```

```python
# Get batch summary
summary = batch_summary(data)

# Find best batch
best_batch = rank_batches(summary)

# Top performers (consumption >= 70)
top_students = find_top_performers(data, "consumption", 70)

# At-risk students (consumption < 20)
risk_students = find_at_risk_students(data, "consumption", 20)

# Print Results
print("Best Batch:", best_batch)

print("\nBatch Summary:")
for batch in summary:
    print(batch,
          "- Avg Consumption:", round(summary[batch]["avg_consumption"], 1),
          "%, Avg Attendance:", round(summary[batch]["avg_attendance"], 1), "%")

print("\nTop Performers:", top_students)

print("\nAt-Risk Students:", risk_students)
```
