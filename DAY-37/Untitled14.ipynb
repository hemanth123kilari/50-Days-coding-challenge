```python
def calculate_total(marks):
    total = 0
    for m in marks:
        total += m
    return total

marks = [85, 90, 80]
total = calculate_total(marks)
print("Total:", total)
```

```python
def get_marks():
    return [85, 90, 80]

def calculate_total(marks):
    return sum(marks)

def calculate_average(total):
    return total / 3

def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"

# Main
marks = get_marks()
total = calculate_total(marks)
avg = calculate_average(total)
grade = get_grade(avg)

print("Marks:", marks)
print("Total:", total)
print("Average:", avg)
print("Grade:", grade)
```
