```python
def calculate_premium(policy_type, age, claims):
    # Base premium setup
    if policy_type == "health":
        base_premium = 8000
    elif policy_type == "vehicle":
        base_premium = 12000
    elif policy_type == "life":
        base_premium = 10000
    else:
        return "Invalid policy type"

    # Start with base premium
    final_premium = base_premium

    # Age adjustment
    if age < 25:
        final_premium += 0.20 * base_premium
    elif age > 45:
        final_premium += 0.15 * base_premium

    # Claims adjustment
    if claims > 2:
        final_premium += 0.25 * base_premium
    elif claims == 0:
        final_premium -= 0.10 * base_premium

    return int(final_premium)
```

```python
name = input("Enter customer name: ")
age = int(input("Enter age: "))
policy_type = input("Enter policy type (health/vehicle/life): ").lower()
claims = int(input("Enter number of claims: "))
```

```python
result = calculate_premium(age, policy_type, claims)
```

```python

if isinstance(result, str):
    print(result)
else:
    print("\n--- Insurance Details ---")
    print("Customer Name:", name)
    print("Final Premium: ₹", result)
```
